```python
import matplotlib.pyplot as plt  
import numpy as np 

x = np.arange(0.0, 2.0, 0.01)
y = np.sin(2*np.pi*x)
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
